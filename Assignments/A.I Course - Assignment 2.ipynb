```python
from sklearn import preprocessing
  
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels for sex - male, female
df['Sex']= label_encoder.fit_transform(df['Sex'])
```

```python
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
df['Cholesterol']= label_encoder.fit_transform(df['Cholesterol'])
  
```

```python
df['Drug'].value_counts()
```

```python
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
df['Drug']= label_encoder.fit_transform(df['Drug'])
  
df['Drug'].unique()
```

```python
df.info()
```

```python
df.head()
```

```python
#splitting the data
X = df.drop("Drug", axis = 1).values
Y = df['Drug'].values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=21)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
     
```

```python
# Initializing the seq model
model = Sequential()
#input layer
model.add(Dense(8, input_dim=5, kernel_initializer='uniform', activation='relu'))
#second hidden layer
model.add(Dense(32,kernel_initializer='uniform',activation='relu'))
#third hidden layer
model.add(Dense(16,kernel_initializer='uniform', activation='relu'))
#fourth hidden layer
model.add(Dense(26,kernel_initializer='uniform', activation='relu'))
#output layer
model.add(Dense(1,activation='softmax'))
     
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(xtrain,ytrain,epochs=100,batch_size=15,validation_data=(xtest,ytest))
```

```python
ypred = model.predict(xtest)
```

```python
comp = pd.DataFrame(ytest)  # Creating a dataframe
comp.columns = ['Actual Value']  # Changing the column name
comp['Predicted values'] = ypred  # Creating a column based onth prediction
comp
```
